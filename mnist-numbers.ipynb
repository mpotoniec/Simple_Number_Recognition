# Recognition of simple images with numbers with artifical neural network

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
print('Setup Complite')
```

# Loading mnist dataset from keras database

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
y_train.shape
```

# Data normalization

```python
X_train = X_train / 255
X_test = X_test / 255
```

# Defining and fitting model

```python
def show_image(X_train, y_train, index):
    plt.imshow(X_train[index])
    plt.xlabel('Actual result is ' + str(y_train[index]))
```

# Defining and fitting cnn model

```python
X_train.shape
```

```python
X_train = X_train.reshape(-1,28,28,1)
X_train.shape
```

```python
X_test = X_test.reshape(-1,28,28,1)
X_test.shape
```

```python
cnn_model = keras.Sequential([
    keras.layers.Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)),
    keras.layers.MaxPooling2D((2, 2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(10, activation = 'sigmoid')
])

cnn_model.compile(
optimizer = 'adam',
loss = 'sparse_categorical_crossentropy',
metrics = ['accuracy'])
```

```python
cnn_model.fit(X_train, y_train, epochs = 20)
```

```python
cnn_model.evaluate(X_test, y_test)
```

# Model have achieved 99% accuracy on last epoch and 98% accuracy o test data
